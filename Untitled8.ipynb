```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


df = pd.read_csv("him.csv")


X = df[['Retweets', 'Hour']]
Y = df['Likes']


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)


print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Mean Squared Error:", mean_squared_error(Y_test, Y_pred))
print("R-squared Score:", r2_score(Y_test, Y_pred))
```

Time Series Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("him.csv", parse_dates=["Timestamp"])
df.set_index("Timestamp", inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Likes"], label="Likes Over Time", color="blue")
plt.xlabel("Time")
plt.ylabel("Likes")
plt.title("Time-Series Data: Likes Over Time")
plt.legend()
plt.show()

df["Moving_Avg"] = df["Likes"].rolling(window=7).mean()
plt.figure(figsize=(12, 6))
plt.plot(df.index, df["Likes"], label="Original Data", alpha=0.5)
plt.plot(df.index, df["Moving_Avg"], label="Trend (7-day Avg)", color="red")
plt.legend()
plt.title("Trend Detection")
plt.show()

decomposition = seasonal_decompose(df["Likes"], model='additive', period=30)
decomposition.plot()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("him.csv", parse_dates=["Timestamp"])
df.set_index("Timestamp", inplace=True)

decomposition = seasonal_decompose(df["Likes"], model='additive', period=30)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df.index, df["Likes"], label="Original", color="blue")
plt.legend()

plt.subplot(412)
plt.plot(df.index, trend, label="Trend", color="red")
plt.legend()

plt.subplot(413)
plt.plot(df.index, seasonal, label="Seasonality", color="green")
plt.legend()

plt.subplot(414)
plt.plot(df.index, residual, label="Residuals", color="gray")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("him.csv")
X = df[['Likes', 'Retweets', 'Hour']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)  # Sum of squared distances to closest centroid

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='blue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
optimal_k = 3  # Adjust based on the elbow method graph
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel("Likes")
plt.ylabel("Retweets")
plt.title("K-Means Clustering Visualization")
plt.colorbar(label="Cluster")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("him.csv")

X = df[['Likes', 'Retweets', 'Hour']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

optimal_k = 3  # Adjust based on the elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel("Likes (Standardized)")
plt.ylabel("Retweets (Standardized)")
plt.title("K-Means Clustering Visualization")
plt.colorbar(label="Cluster")
plt.show()
```
