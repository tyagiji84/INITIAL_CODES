```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("him.csv", parse_dates=["Timestamp"])
df.set_index("Timestamp", inplace=True)

decomposition = seasonal_decompose(df["Likes"], model='additive', period=30)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(df.index, df["Likes"], label="Original", color="blue")
plt.legend()

plt.subplot(412)
plt.plot(df.index, trend, label="Trend", color="red")
plt.legend()

plt.subplot(413)
plt.plot(df.index, seasonal, label="Seasonality", color="green")
plt.legend()

plt.subplot(414)
plt.plot(df.index, residual, label="Residuals", color="gray")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("him.csv")
X = df[['Likes', 'Retweets', 'Hour']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
k_values = range(1, 10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)  # Sum of squared distances to closest centroid

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='blue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
optimal_k = 3  # Adjust based on the elbow method graph
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.5)
plt.xlabel("Likes")
plt.ylabel("Retweets")
plt.title("K-Means Clustering Visualization")
plt.colorbar(label="Cluster")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


df = pd.read_csv("him.csv")


X = df[['Retweets', 'Likes', 'Hour', 'Platform', 'Country']]
Y = df['Sentiment']

categorical_features = ['Platform', 'Country']
numerical_features = ['Retweets', 'Likes', 'Hour']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

X_processed = preprocessor.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_processed, Y, test_size=0.2, random_state=42)
```

```python
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier()
}

for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"{name} Accuracy: {accuracy:.4f}")
    print(classification_report(Y_test, Y_pred))
    print("-" * 50)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("him.csv")

X = df[['Retweets', 'Likes', 'Hour', 'Platform', 'Country']]
Y = df['Sentiment']

categorical_features = ['Platform', 'Country']
numerical_features = ['Retweets', 'Likes', 'Hour']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

X_processed = preprocessor.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_processed, Y, test_size=0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_model = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_model = grid_search.best_estimator_
Y_pred = best_model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print("Optimized Model Accuracy:", accuracy)
print(classification_report(Y_test, Y_pred))
```

```python
import pandas as pd
import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Load dataset
df = pd.read_csv("him.csv")

# Remove missing values in text column
df.dropna(subset=["Text"], inplace=True)

# Tokenization
df["Tokens"] = df["Text"].apply(word_tokenize)

# Stopword Removal
stop_words = set(stopwords.words("english"))

def remove_stopwords(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

df["Cleaned_Tokens"] = df["Tokens"].apply(remove_stopwords)

# Stemming
stemmer = PorterStemmer()

def apply_stemming(tokens):
    return [stemmer.stem(word) for word in tokens]

df["Stemmed_Tokens"] = df["Cleaned_Tokens"].apply(apply_stemming)

# Lemmatization
def apply_lemmatization(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc]

df["Lemmatized_Text"] = df["Text"].apply(apply_lemmatization)

# Display processed data
print(df[["Text", "Tokens", "Cleaned_Tokens", "Stemmed_Tokens", "Lemmatized_Text"]].head())
```

```python
import pandas as pd
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required NLTK resources
nltk.download('punkt')
nltk.download('vader_lexicon')

# Load dataset
df = pd.read_csv("him.csv")
df.dropna(subset=["Text"], inplace=True)  # Remove missing values

# Sentiment Analysis using TextBlob
def textblob_sentiment(text):
    score = TextBlob(text).sentiment.polarity
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df["Sentiment"] = df["Text"].apply(textblob_sentiment)

# Sentiment Distribution Visualization
plt.figure(figsize=(8, 5))
sns.countplot(x=df["Sentiment"], palette="coolwarm")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()

# Generate Word Cloud for Entire Text Dataset
text = " ".join(df["Text"])
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Text Data")
plt.show()

# Generate Word Clouds for Each Sentiment
sentiments = ["Positive", "Negative", "Neutral"]
for sentiment in sentiments:
    filtered_text = " ".join(df[df["Sentiment"] == sentiment]["Text"])
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(filtered_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Word Cloud for {sentiment} Sentiment")
    plt.show()
```
